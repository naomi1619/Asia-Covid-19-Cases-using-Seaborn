```python
import numpy as np
import pandas as pd
import plotly.express as px
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('AsiaCases_.csv')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
sns.heatmap(df.isnull())
```

```python

plt.figure(figsize = (10, 8))
sns.heatmap(df.corr(), annot = True)
```

```python
# Checking Total cases by country
plt.figure(figsize = (10, 15))
plt.title("TotalCases by Country", fontsize = 18)
sns.barplot(data = df, x = "TotalCases", y = "Country")
```

```python
# Checking Total deaths by country
plt.figure(figsize = (10, 15))
plt.title("TotalDeaths by Country", fontsize = 18)
sns.barplot(data = df, x = "TotalDeaths", y = "Country")
```

```python
# Checking Total recoverd by country
plt.figure(figsize = (10, 15))
plt.title("TotalRecovered by Country", fontsize = 18)
sns.barplot(data = df, x = "TotalRecovered", y = "Country")
```

```python
# Checking Total tests by country
plt.figure(figsize = (10, 15))
plt.title("TotalTests by Country", fontsize = 18)
sns.barplot(data = df, x = "TotalTests", y = "Country")
```

```python
# Checking Total population by country
plt.figure(figsize = (10, 15))
plt.title("TotalPopulation by Country", fontsize = 18)
sns.barplot(data = df, x = "TotalPopulation", y = "Country")
```

```python
plt.figure(figsize = (12, 18))
fig = px.bar(df, x = "TotalCases", y = "Country", color = "TotalPopulation")
fig.show()
```
